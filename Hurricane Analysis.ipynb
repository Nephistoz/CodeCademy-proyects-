1. Hurricanes, also known as cyclones or typhoons, are one of the most powerful forces of nature on Earth. For some of the hurricanes, damage data was not recorded (`"Damages not recorded"`), while the rest are written in the format `"Prefix-B/M"`, where `B` stands for billions (`1000000000`) and `M` stands for millions (`1000000`).
   
   Write a function that returns a new list of updated damages where the recorded data is converted to float values and the missing data is retained as `"Damages not recorded"`.
   
   Test your function with the data stored in `damages`.

````python
# damages (USD($)) of hurricanes
damages = ['Damages not recorded', '100M', 'Damages not recorded', '40M',
          '27.9M', '5M', 'Damages not recorded', '306M', '2M', '65.8M',
          '326M', '60.3M', '208M', '1.42B', '25.4M', 'Damages not recorded',
          '1.54B', '1.24B', '7.1B', '10B', '26.5B', '6.2B', '5.37B', '23.3B',
          '1.01B', '125B', '12B', '29.4B', '1.76B', '720M', '15.1B', '64.8B',
          '91.6B', '25.1B']

# 1
# Update Recorded Damages
conversion = {"M": 1000000,
             "B": 1000000000}

def damage_updated(damages_list):
    updated_records = []
    for item in damages_list:
        if item == "Damages not recorded":
            updated_records.append(item)
        else:
            # Detectar si el valor contiene "M" o "B" y realizar la conversiÃ³n correspondiente
            if "M" in item:
                numeric_value = float(item.replace("M", "")) * conversion["M"]
            elif "B" in item:
                numeric_value = float(item.replace("B", "")) * conversion["B"]
            else:
                numeric_value = float(item)
                
            updated_records.append(numeric_value)

    return updated_records
# test function by updating damages
print(damage_updated(damages))
````

5. Write a function that finds the area affected by the most hurricanes, and how often it was hit.

   Test your function on your affected area dictionary.

````python
# 5
# Calculating Maximum Hurricane Count
def most_area_affected(hurricane_dictionary):
    count_area_affected= {}
    
    for values in hurricane_dictionary.values():
        for area in values["Areas Affected"]:
            if area not in count_area_affected:
                count_area_affected[area] = 1
            else:
                count_area_affected[area] +=1
                
 # find most frequently affected area and the number of hurricanes involved in 
    most_affected_area= None
    most_count = 0
    
    for area, count in count_area_affected.items():
        if count > most_count:
            most_affected_area = area
            most_count = count  
                      
    return {most_affected_area: most_count}

print(most_area_affected(hurricanes_data_dictionary))

# find most frequently affected area and the number of hurricanes involved in 
````

   ```py
   mortality_scale = {0: 0,
   1: 100,
   2: 500,
   3: 1000,
   4: 10000}
   ```
   
   For example, a hurricane with a `1` mortality rating would have resulted in greater than `0` but less than or equal to `100` deaths. A hurricane with a `5` mortality would have resulted in greater than `10000` deaths.
   
   Store the hurricanes in a new dictionary where the keys are the mortaility ratings and the values are lists containing a dictionary for each hurricane that falls into that mortality rating.
   
   Test your function on your hurricane dictionary.

````python
# 7
# Rating Hurricanes by Mortality
mortality_scale = {
0: 0,
1: 100,
2: 500,
3: 1000,
4: 10000,
}

def rating_hurricanes_by_mortality(hurricane_dictionary):
    mortality_dictionary = {
    0: [],
    1: [],
    2: [],
    3: [],
    4: [],
    5: []    
    }
    
    for hurricane,values in hurricane_dictionary.items():
        match values["Deaths"]:
            case _ if values["Deaths"] > mortality_scale[4]:
                mortality_dictionary[5].append(values) 
            case _ if mortality_scale [3] <= values["Deaths"] <= mortality_scale[4]:
                mortality_dictionary[4].append(values)
            case _ if mortality_scale [2] <= values["Deaths"] <= mortality_scale[3]:
                mortality_dictionary[3].append(values) 
            case _ if mortality_scale [1] <= values["Deaths"] <= mortality_scale[2]:
                mortality_dictionary[2].append(values)   
            case _ if mortality_scale [0] <= values["Deaths"] <= mortality_scale[1]:
                mortality_dictionary[1].append(values) 
            case _ if values["Deaths"] == 0:
                mortality_dictionary[0] = hurricane            
    return mortality_dictionary    
                
                
            
rating_hurricanes_by_mortality(hurricanes_data_dictionary)

# categorize hurricanes in new dictionary with mortality severity as key
````

9. Lastly, you want to rate hurricanes according to how much damage they cause.

   Write a function that rates hurricanes on a damage scale according to the following ratings, where the key is the rating and the value is the upper bound of damage for that rating.
   ```py
   damage_scale = {0: 0,

   3: 10000000000,
   4: 50000000000}
   ```
   
   For example, a hurricane with a `1` damage rating would have resulted in damages greater than `0` USD but less than or equal to `100000000` USD. A hurricane with a `5` damage rating would have resulted in damages greater than `50000000000` USD (talk about a lot of money).

   Test your function on your hurricane dictionary.

````python
# 9
# Rating Hurricanes by Damage
damage_scale = {0: 0,
1: 100000000,
2: 1000000000,
3: 10000000000,
4: 50000000000
}
conversion = {"M": 1000000,
             "B": 1000000000}

def rating_hurricanes_by_damages(hurricane_dictionary):
    damages_dictionary = {
    "Damage not recorded or 0": [],
    1: [],
    2: [],
    3: [],
    4: [],
    5: []    
        }
    
    ## Conversion
    for hurricane,values in hurricane_dictionary.items():
        if "M" in hurricane_dictionary[hurricane]["Damages"]:
            numeric_value = float(hurricane_dictionary[hurricane]["Damages"].replace("M", "")) * conversion["M"]
        elif "B" in hurricane_dictionary[hurricane]["Damages"]:
            numeric_value = float(hurricane_dictionary[hurricane]["Damages"].replace("B", "")) * conversion["B"]
        elif hurricane_dictionary[hurricane]["Damages"] == "Damages not recorded":
            numeric_value = 0
    
        match numeric_value:
            case _ if numeric_value > damage_scale[4]:
                damages_dictionary[5].append(values) 
            case _ if damage_scale [3] <= numeric_value <= damage_scale[4]:
                damages_dictionary[4].append(values)
            case _ if damage_scale [2] <= numeric_value <= damage_scale[3]:
                damages_dictionary[3].append(values) 
            case _ if damage_scale [1] <= numeric_value <= damage_scale[2]:
                damages_dictionary[2].append(values)   
            case _ if 0 < numeric_value <= damage_scale[1]:
                damages_dictionary[1].append(values) 
            case _ if numeric_value == 0:
                damages_dictionary["Damage not recorded or 0"].append(values)            
    return damages_dictionary 

rating_of_damage = rating_hurricanes_by_damages(hurricanes_data_dictionary)
print(rating_of_damage["Damage not recorded or 0"])
# categorize hurricanes in new dictionary with damage severity as key
````
